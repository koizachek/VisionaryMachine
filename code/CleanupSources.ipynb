```python
import pandas as pd, numpy as np
```

```python
s = pd.read_csv('search_results.csv')
s_2030 = pd.read_csv('search_results_2030.csv')
s_2040 = pd.read_csv('search_results_2040.csv')
s_2050 = pd.read_csv('search_results_2050.csv')
```

```python
# concatenate the dataframes into a single dataframe
merged_df = pd.concat([s, s_2030, s_2040, s_2050], ignore_index=True)
merged_df
```

```python
# remove duplicates based on the 'title' column
merged_df.drop_duplicates(subset=['Title'], keep='first', inplace=True)
```

```python
merged_df
```

```python
# filter the dataframe based on the 'title' column
mask = merged_df['Title'].str.contains('scenario|Scenario|Scenarios|scenarios')
merged_df = merged_df[mask]
```
